# Data Transformation


```python
#DataFrame of aggregated transactions

path1 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\aggregated\transaction\country\india\state\\"

agggregated_trans_list = os.listdir(path1)

header1 = {'State':[], 'Year': [], 'Quarter' : [], 'Transaction_type' : [], 'Transaction_count' : [], 'Transaction_amount' : []}

for state in agggregated_trans_list:
    current_state = path1 + state + "/"
    aggregated_year_list = os.listdir(current_state)
    
    for year in aggregated_year_list:
        current_year = current_state + year + "/"
        aggregated_file_list = os.listdir(current_year)
        
        for file in aggregated_file_list:
            current_file = current_year + file
            data = open(current_file, 'r')
            F = json.load(data)
            
            for i in F['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                header1['Transaction_type'].append(name)
                header1['Transaction_count'].append(count)
                header1['Transaction_amount'].append(amount)
                header1['State'].append(state)
                header1['Year'].append(year)
                header1['Quarter'].append(int(file.strip('.json')))
                
df_agg_transaction = pd.DataFrame(header1)
```

```python
#Dataframe of aggregated user

path2 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\aggregated\user\country\india\state\\"


aggregated_user_list = os.listdir(path2)

header2 = {'State':[], 'Year': [], 'Quarter' : [], 'Brand' : [], 'Count' : [], 'Percentage' : []}

for state in aggregated_user_list:
    current_state = os.path.join(path2, state) 
    aggregated_year_list = os.listdir(current_state)
    
    for year in aggregated_year_list:
        current_year = os.path.join(current_state, year)
        aggregated_file_list = os.listdir(current_year)
        
        for file in aggregated_file_list:
            current_file = os.path.join(current_year, file)
            data = open(current_file, 'r')
            F1 = json.load(data)
            try:
                for i in F1['data']['usersByDevice']:
                    brand_name = i['brand']
                    count = i['count']
                    percentage = i['percentage']
                    header2["Brand"].append(brand_name)
                    header2["Count"].append(count)
                    header2["Percentage"].append(percentage)
                    header2["State"].append(state)
                    header2["Year"].append(year)
                    header2["Quarter"].append(int(file.strip('.json')))
            except:
                pass

df_aggregated_user = pd.DataFrame(header2)
```

```python
#Dataframe of Mapped Transactions

path3 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\map\transaction\hover\country\india\state"

map_list = os.listdir(path3)

header3 = {'State':[], 'Year': [], 'Quarter' : [], 'District': [], 'Count':[], 'Amount':[]}

for state in map_list:
    current_state = os.path.join(path3,state)
    map_year = os.listdir(current_state)
    
    for year in map_year:
        current_year = os.path.join(current_state,year)
        map_file = os.listdir(current_year)
        
        for file in map_file:
            current_file = os.path.join(current_year,file)
            data = open(current_file,'r')
            F2 = json.load(data)
            try:
                for i in F2['data']['hoverDataList']:
                    district = i['name']
                    count = i['metric'][0]['count']
                    amount = i['metric'][0]['amount']
                    header3['District'].append(district)
                    header3['Count'].append(count)
                    header3['Amount'].append(amount)
                    header3['State'].append(state)
                    header3["Year"].append(year)
                    header3["Quarter"].append(int(file.strip('.json')))
            except:
                pass
df_map_list = pd.DataFrame(header3)
```

```python
#Dataframe of Mapped user

path4 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\map\user\hover\country\india\state\\"

map_user_list = os.listdir(path4)

header4 = {"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}

for state in map_user_list:
    cur_state = path4 + state + "/"
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state + year + "/"
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            F3 = json.load(data)
            
            for i in F3["data"]["hoverData"].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appOpens = i[1]['appOpens']
                header4["District"].append(district)
                header4["RegisteredUser"].append(registereduser)
                header4["AppOpens"].append(appOpens)
                header4['State'].append(state)
                header4['Year'].append(year)
                header4['Quarter'].append(int(file.strip('.json')))
                
df_map_user = pd.DataFrame(header4)
```

```python
#Dataframe of top transactions

path5 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\top\transaction\country\india\state\\"

top_transaction = os.listdir(path5)
header5 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_transaction:
    current_state = path5 + state + "/"
    top_year = os.listdir(current_state)
    
    for year in top_year:
        current_year = current_state + year + "/"
        top_file = os.listdir(current_year)
        
        for file in top_file:
            current_file = current_year + file
            data = open(current_file, 'r')
            F4 = json.load(data)
            
            for i in F4['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                header5['Pincode'].append(name)
                header5['Transaction_count'].append(count)
                header5['Transaction_amount'].append(amount)
                header5['State'].append(state)
                header5['Year'].append(year)
                header5['Quarter'].append(int(file.strip('.json')))
df_top_transaction = pd.DataFrame(header5)

```

```python
#Dataframe of top users

path6 = r"C:\Users\Vijai\Documents\PhonepePulse\Data\data\top\user\country\india\state\\"
top_user = os.listdir(path6)

header6 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'RegisteredUsers' : [] }

for states in top_user:
    current_state = path6 + states + "/"
    top_states_list = os.listdir(current_state)
    
    for year in top_states_list:
        current_year = current_state + year + "/"
        top_year_list = os.listdir(current_year)
        
        for files in top_year_list:
            current_file = current_year + files
            data = open(current_file,'r')
            F5 = json.load(data)
            
            for i in F5['data']['pincodes']:
                name = i['name']
                registeredUsers = i['registeredUsers']
                header6['Pincode'].append(name)
                header6['RegisteredUsers'].append(registeredUsers)
                header6['State'].append(states)
                header6['Year'].append(year)
                header6['Quarter'].append(int(file.strip('.json')))
                
df_top_user = pd.DataFrame(header6)
            
```

```python
# Connect to the SQLite database
db_path = 'phonepe_pulse.db'
connection = sqlite3.connect(db_path)
cursor = connection.cursor()

# Create the 'agg_transaction' table if it doesn't exist
create_agg_trans = '''
CREATE TABLE IF NOT EXISTS agg_transactions (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    Transaction_type TEXT,
    Transaction_count INTEGER,
    Transaction_amount REAL
)
'''
cursor.execute(create_agg_trans)

# Iterate through the DataFrame rows and insert them into the table
for i, row in df_agg_transaction.iterrows():
    sql = "INSERT INTO agg_transactions VALUES (?, ?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql, values)

# Commit the changes and close the connection
connection.commit()
connection.close()
```

```python
db_path = 'phonepe_pulse.db'
connection = sqlite3.connect(db_path)
cursor = connection.cursor()


#Creating aggregated_user table

# Drop the existing 'agg_users' table if it exists
cursor.execute("DROP TABLE IF EXISTS agg_users")

# Create the 'agg_users' table with the correct columns
create_agg_users_table_query = '''
CREATE TABLE agg_users (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    Brands TEXT,
    Count INTEGER,
    Percentage REAL
)
'''
cursor.execute(create_agg_users_table_query)

#Iterate through DF and insert them into the table
for i,row in  df_aggregated_user.iterrows():
    sql = "INSERT INTO agg_users VALUES (?, ?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql,values)
    
connection.commit()

    

#create map_list table

# Drop the existing 'map_transactions' table if it exists
cursor.execute("DROP TABLE IF EXISTS map_transactions")

# Create the 'map_transactions' table with the correct columns
create_map_transactions_table_query = '''
CREATE TABLE map_transactions (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    District TEXT,
    Count INTEGER,
    Amount REAL
)
'''
cursor.execute(create_map_transactions_table_query)

#iterate 
for i,row in  df_map_list.iterrows():
    sql = "INSERT INTO map_transactions VALUES (?, ?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql,values)
    
    
connection.commit()

    
#Create map_user table
cursor.execute("DROP TABLE IF EXISTS map_users")

# Create the 'map_users' table with the correct columns
create_map_users_table_query = '''
CREATE TABLE map_users (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    District TEXT,
    Registered_user INTEGER,
    App_opens INTEGER
)
'''
cursor.execute(create_map_users_table_query)

# Insert data into 'map_users' table
for i, row in df_map_user.iterrows():
    sql = "INSERT INTO map_users VALUES (?, ?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql, values)

# Commit the changes after all operations are done
connection.commit()

    
    
#create top_transaction table

# Drop the existing 'top_transaction' table if it exists
cursor.execute("DROP TABLE IF EXISTS top_transaction")

# Create the 'top_transaction' table with the correct columns
create_top_transaction_table_query = '''
CREATE TABLE top_transaction (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    Pincode TEXT,
    Transaction_count INTEGER,
    Transaction_amount REAL
)
'''
cursor.execute(create_top_transaction_table_query)

# Insert data into 'top_transaction' table
for i, row in df_top_transaction.iterrows():
    sql = "INSERT INTO top_transaction VALUES (?, ?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql, values)

# Commit the changes after all operations are done
connection.commit()
    
    
#Creating top_users tables

# Drop the existing 'top_user' table if it exists
cursor.execute("DROP TABLE IF EXISTS top_user")

# Create the 'top_user' table with the correct columns
create_top_user_table_query = '''
CREATE TABLE top_user (
    State TEXT,
    Year INTEGER,
    Quarter INTEGER,
    Pincode INTEGER,
    Registered_users INTEGER
)
'''
cursor.execute(create_top_user_table_query)

# Insert data into 'top_user' table
for i, row in df_top_user.iterrows():
    sql = "INSERT INTO top_user VALUES (?, ?, ?, ?, ?)"
    values = tuple(row)
    cursor.execute(sql, values)

# Commit the changes after all operations are done
connection.commit()
connection.close()
```
